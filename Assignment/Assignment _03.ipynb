# Task 1: Inventory Price Calculator


```python
# Task 1: Inventory Price Calculator

# 1. Define the function structure which accepts a list of prices
def calculate_inventory_summary(prices):
    
    # Check if the list is empty to prevent ZeroDivisionError
    if not prices:
        return (0, 0.0)
    
    # 2. Calculate the total cost using the built-in sum() function
    total_cost = sum(prices)

    # 3. Calculate the number of items (length of the list)
    number_of_items = len(prices)

    # 4. Calculate the average cost (Standard division ensures a float result)
    average_cost = total_cost / number_of_items

    # 5. Return the total cost and average cost as a tuple
    return (total_cost, average_cost)

# Example Usage:
input_prices = [120, 250, 399, 150]

# 6. Call the function and print the returned tuple
summary = calculate_inventory_summary(input_prices)
print(summary)
# Expected Output: (919, 229.75)
```

# Task 2: Unique Visitor Tracker


```python

# 1. Define the function structure which accepts a list of visitor IDs
def count_unique_visitors(visitor_list):
    
    # 2. Convert the list of IDs into a set
    # Sets automatically discard duplicate elements
    unique_ids_set = set(visitor_list)

    # 3. Calculate the number of unique elements (the length of the set)
    number_of_unique_visitors = len(unique_ids_set)
    
    # 4. Return the final count
    return number_of_unique_visitors

# Example Usage:
input_visitors = [101, 205, 101, 310, 205, 550]

# 5. Call the function and print the returned result
unique_count = count_unique_visitors(input_visitors)
print(unique_count)
# Expected Output: 4
```

# Task 3: Product Category Counter


```python

# 1. Define the function structure which accepts a list of product names
def count_categories(products):
    
    # 2. Initialize an empty dictionary to store product counts (Category: Count)
    category_counts = {}

    # 3. Loop through each item in the product list
    for product in products:
        
        # 4. Check if the product name is already a key in the dictionary
        if product in category_counts:
            # If the category exists, increment its count by 1
            category_counts[product] += 1
        else:
            # If the category is new, add it with an initial count of 1
            category_counts[product] = 1
            
    # 5. Return the final dictionary containing all category counts
    return category_counts

# Example Usage:
input_products = ["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]

# 6. Call the function and print the returned result
product_summary = count_categories(input_products)
print(product_summary)
# Expected Output: {"Laptop": 3, "Smartphone": 2, "Monitor": 1, "Headphones": 1}
```

# Temperature Analytics

```python
def analyze_temperatures(temp_list):
    high_list = []
    normal_list = []
    THRESHOLD = 38.0
    
    for temp in temp_list:
        # High is strictly greater than 38.0
        if temp > THRESHOLD:
            high_list.append(temp)
        else:
            # Normal is 38.0 and below
            normal_list.append(temp)
            
    return (high_list, normal_list)

# Example Usage:
input_temps = [36.5, 38.0, 39.1, 37.2, 40.5, 38.0, 37.9]
print(analyze_temperatures(input_temps))
```

# Menu Price Lookup

```python
def get_price(menu, item):
    # Check if the item name exists as a key in the dictionary
    if item in menu:
        return menu[item]
    else:
        return "Item not found"

# Example Usage:
menu_data = {"Coffee": 3.50, "Tea": 2.75, "Muffin": 4.00, "Sandwich": 8.50}
print(get_price(menu_data, "Muffin"))
print(get_price(menu_data, "Donut"))
```

# Student Grading System

```python
# Function 1: Calculate Grade based on score
def calculate_grade(score):
    if score >= 85: # Assuming >= 85 is A+ based on example output
        return "A+"
    elif score >= 70:
        return "A"
    elif score >= 60:
        return "A-"
    else:
        return "F"

# Function 2: Grade Students Dictionary
def grade_students(student_dict):
    student_grades = {}
    # Loop through each student and score pair in the dictionary
    for student, score in student_dict.items():
        grade = calculate_grade(score)
        student_grades[student] = grade
    return student_grades

# Example Usage:
input_scores = {"Alice": 85, "Bob": 78, "Charlie": 32, "David": 69}
print(grade_students(input_scores))
```

# Discount Calculator

```python
def apply_discounts(products):
    updated_prices = []
    # products is a list of tuples: (name, price, discount)
    for name, price, discount in products:
        # Calculate discount rate (e.g., 20 / 100 = 0.20)
        discount_rate = discount / 100
        # Calculate final price: price * (1 - discount rate)
        final_price = price * (1 - discount_rate)
        
        # Round the price to two decimal places
        updated_prices.append(round(final_price, 2))
    return updated_prices

# Example Usage:
input_products = [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]
print(apply_discounts(input_products))
```

# Hashtag Generator

```python
def generate_hashtags(sentence):
    # Convert sentence to lowercase and split into words
    words = sentence.lower().split()
    
    hashtags = []
    for word in words:
        # Prepend '#' to each word
        hashtags.append("#" + word)
        
    return hashtags

# Example Usage:
input_sentence = "Al Data Science"
print(generate_hashtags(input_sentence))
```

```python
# Mock Database provided in the assignment
CREDENTIALS = {"alex_s": "securepwd1", "maria_t": "password101", "john_d": "johndoe123"}

def login(username, password):
    
    # 1. Check if the username exists in the mock database keys
    if username in CREDENTIALS:
        # 2. If the username exists, check if the password matches the stored value
        if CREDENTIALS[username] == password:
            return "Login Successful"
        else:
            # Username exists, but password is wrong
            return "Invalid Credentials"
    else:
        # 3. Username does not exist
        return "Invalid Credentials"

# Example Usage:
print(login("maria_t", "password101")) # Successful
print(login("alex_s", "securepwd12"))  # Failed (Wrong Password)
```

# Shopping Cart Bill Calculator

```python
def calculate_total(cart, prices):
    total_bill = 0.0
    
    # Iterate through each item in the shopping cart list
    for item in cart:
        # Check if the item exists in the price dictionary
        if item in prices:
            # Add the price of the item to the total bill
            total_bill += prices[item]
            
    # Round the total bill to two decimal places for currency
    return round(total_bill, 2)

# Example Usage:
input_cart = ["Milk", "Bread", "Milk", "Cheese", "Bread"]
price_dict = {"Milk": 3.00, "Bread": 2.50, "Eggs": 4.00, "Cheese": 5.00}
print(calculate_total(input_cart, price_dict))
```
