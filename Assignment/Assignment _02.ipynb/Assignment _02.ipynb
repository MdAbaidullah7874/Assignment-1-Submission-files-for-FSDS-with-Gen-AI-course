# Task 2: Count Word Frequency in Customer Reviews

```python


# 1. Define the input review text
review_text = "This product is good and this service is excellent" # Input review

# 2. Convert the entire text to lowercase 
# Converting to lowercase to remove case sensitivity.
lowercase_text = review_text.lower() 

# 3. Split the text into a list of individual words 
# Splitting the text into a list of words.
word_list = lowercase_text.split() 

# 4. Initialize an empty dictionary to store word counts
# Dictionary to store each word's frequency (key: word, value: count).
word_counts = {}

# 5. Iterate through the word list and count frequencies
# Looping through the list of words.
for word in word_list:
    # Check if the word is already in the dictionary
    if word in word_counts:
        # If yes, increment the count by 1
        word_counts[word] += 1
    else:
        # If no, add the word to the dictionary with a count of 1
        word_counts[word] = 1

# 6. Print the final word counts dictionary
print("Word Counts:")
print(word_counts)
```

# # Task 3: Password Strength Checker

```python
# 1. Define the input password
password = "Hello@123" 

# 2. Initialize four boolean flags to False

has_upper = False
has_lower = False
has_digit = False
has_special = False


# 3. Loop through each character in the password
for char in password:
    if char.isupper():
        has_upper = True #  if_upper=True
    elif char.islower():
        has_lower = True # is_lower=True
    elif char.isdigit():
        has_digit = True # is_digit=True
    # A character is 'special' if it is NOT an alphabet AND NOT a number.
    elif not char.isalnum(): 
        has_special = True # is_special=True

# 4. Check if all four flags are True using the 'and' operator
if has_upper and has_lower and has_digit and has_special:
    strength_status = "Strong Password"
else:
    strength_status = "Weak Password"

# 5. Print the final result
print(strength_status)
```

# Task 4: Username Generator


```python
# Task 4: Username Generator

import random

# 1. Define the input name
full_name = "Mamun Malitha"

# 2. Convert to lowercase and split the name into parts
name_parts = full_name.lower().split()

# Check if the name has at least two parts
if len(name_parts) >= 2:
    first_name = name_parts[0]
    last_name = name_parts[-1]
    
    # 3. Get the first three letters of the first name
    # Slicing from index 0 up to (but not including) 3
    first_part = first_name[0:3] 
    
    # 4. Get the last three letters of the last name
    # Negative indexing starts counting from the end
    last_part = last_name[-3:]
    
    # 5. Generate a random two-digit number (10 to 99)
    random_number = random.randint(10, 99)
    
    # 6. Combine all parts to form the username
    # Convert the random number to a string before combining
    username = first_part + last_part + str(random_number)
    
    # 7. Print the generated username
    print(username)

else:
    print("Invalid name format. Please provide a full name with at least two parts.")
```

# Task 5: SMS Word Counter



```python


# 1. Define the input message
message = "This is a demo message."

# 2. Define the SMS character limit constant
SMS_LIMIT = 160


# 3. Calculate the number of characters
#    len() counts all characters, including spaces and punctuation
character_count = len(message)

# 4. Calculate the number of words
#    message.split() creates a list of words, and len() counts them.
word_list = message.split()
word_count = len(word_list)


# 5. Check the status against the SMS limit
if character_count > SMS_LIMIT:
    status = "Exceeded Limit"
else:
    status = "Within Limit"

# 6. Print the required output format using f-string
print(f"Words = {word_count}, Characters = {character_count}, Status = {status}")
```

# Task 6: Remove Duplicates from Contact List


```python


# 1. Define the input string of comma-separated names
contact_list_input = "Rahim, Karim, Rahim, Sultana, Karim"

# 2. Split the string into a list of names using the comma as a separator
name_list = contact_list_input.split(',')

# 3. Clean up the names: remove leading/trailing spaces from each name
#    We use a loop to iterate through all names and strip() method to clean them.
cleaned_names = []
for name in name_list:
    cleaned_names.append(name.strip())

# 4. Convert the list of cleaned names to a set to automatically remove duplicates
unique_names_set = set(cleaned_names)

# 5. Convert the set back to a list (Sets do not preserve order)
unique_names_list = list(unique_names_set)

# 6. Join the unique names back into a comma-separated string with a space after the comma
output_string = ", ".join(unique_names_list)

# 7. Print the final result
print(output_string)
```

# Task 7: Detect Suspicious Words in Text (Moderation)


```python


# 1. Define the input text and the list of banned words
input_text = "This is a scam offer"
banned_words_list = ['spam', 'scam', 'fake']

# 2. Pre-process the input text: convert to lowercase and split into a list of words
text_words_list = input_text.lower().split()

# 3. Convert both the text words and the banned list into Sets
#    This conversion is necessary for the efficient intersection operation
text_words_set = set(text_words_list)
banned_words_set = set(banned_words_list)

# 4. Find the common words using the intersection operator (&)
#    If a suspicious word is present in the text, it will appear in this new set
suspicious_words_found = text_words_set & banned_words_set

# 5. Check if the intersection set is NOT empty
#    If the length of the set is greater than 0, a banned word was found
if suspicious_words_found:
    print("Warning: Suspicious content detected")
else:
    print("Content is clean")
```

# Task 8: Invoice Code Formatter


```python
# Task 8: Invoice Code Formatter

# 1. Define the integer input invoice number
invoice_number = 23

# 2. Define the required prefix string
PREFIX = "INV"

# 3. Define the desired width for the padded number
WIDTH = 5

# 4. Format the number with leading zeros using an f-string
#    The ':05d' specification means: pad with zeros (0) to a total width of 5 digits (5d).
padded_number_str = f"{invoice_number:05d}"

# 5. Concatenate the prefix and the padded number string
invoice_code = PREFIX + padded_number_str

# 6. Print the final formatted code
print(invoice_code)
```

# Task 9: Count Hashtags and Mentions (Social Media Analytics)


```python
# Task 9: Count Hashtags and Mentions (Social Media Analytics)

# 1. Define the input social media post
social_media_post = "Check out #AIWorkshop and follow @inceptionbd"

# 2. Initialize counters for hashtags and mentions
hashtag_count = 0
mention_count = 0

# 3. Split the post into a list of words (tokens)
post_words = social_media_post.split()

# 4. Loop through each word in the post
for word in post_words:
    # 5. Check if the word starts with a hashtag symbol (#)
    if word.startswith("#"):
        hashtag_count += 1
    # 6. Check if the word starts with a mention symbol (@)
    elif word.startswith("@"):
        mention_count += 1

# 7. Print the final counts in the required format
print(f"Hashtags = {hashtag_count}, Mentions = {mention_count}")
```

# Task 10: Word Reversal for Data Encryption


```python
# Task 10: Word Reversal for Data Encryption

# 1. Define the input sentence
input_sentence = "AI is fun"

# 2. Split the sentence into a list of words using spaces
words_list = input_sentence.split()

# 3. Initialize a list to store the reversed words
reversed_words = []

# 4. Loop through each word in the original list
for word in words_list:
    # 5. Reverse the current word using string slicing [::-1]
    #    This slicing trick takes the string backward, effectively reversing it.
    reversed_word = word[::-1]
    
    # Add the reversed word to the new list
    reversed_words.append(reversed_word)

    # 6. Join the reversed words back into a single sentence using spaces
output_sentence = " ".join(reversed_words)

# 7. Print the final result
print(output_sentence)
```
